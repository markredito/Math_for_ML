## Functions in Python

```python
# Simple function f(x) = x^2

def f(x):
    return x ** 2

print(f(3))
```

```python
# derivative

def dfdx(x):
    return 2*x

print(dfdx(3))
```

```python
# apply function to each element in array using numpy
import numpy as np
x_array = np.array([1, 2, 3])

print("X: \n", x_array)
print("f(x) = x**2: \n", f(x_array))
print("f`(x) = 2x: \n", dfdx(x_array))
```

Apply function `f` and `dfdx` to a larger sized array.

```python
import matplotlib.pyplot as plt

# Output of plotting commands is displayed inline within the Jupyter notebook.
%matplotlib inline

def plot_f1_and_f2(f1, f2=None, x_min=-5, x_max=5, label1="f(x)", label2="f'(x)"):
    x = np.linspace(x_min, x_max,100)

    # Setting the axes at the centre.
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    plt.plot(x, f1(x), 'r', label=label1)
    if not f2 is None:
        # If f2 is an array, it is passed as it is to be plotted as unlinked points.
        # If f2 is a function, f2(x) needs to be passed to plot it.        
        if isinstance(f2, np.ndarray):
            plt.plot(x, f2, 'bo', markersize=3, label=label2,)
        else:
            plt.plot(x, f2(x), 'b', label=label2)
    plt.legend()

    plt.show()
    
plot_f1_and_f2(f, dfdx)
```

Define mathematical expression $2x^2 - xy$ using symbolic expression.

```python
# List of symbols.
x, y = symbols ('x y')
#Definition of the expression.
expr = 2 * x**2 - x * y
expr
```

This lets you perform various manipulations with this expression: add, subtract some terms, multiply etc.

```python
expr_manip = x * (expr + x * y + x**3)
expr_manip
```

```python
# Lets try adding the function x3 + y2
expr_manip2 = x**3 + y**2 * (expr)
expr_manip2
```

```python
# Expand expression
expand(expr_manip)
```

```python
# Expand expr_manip2
expand(expr_manip2)
```

```python
# Factorize
factor(expr_manip)
```

```python
# Substitute valuex in expression:
expr.evalf(subs={x:-1, y:2})
```

```python
# Substitute values in expression expr_manip2
expr_manip2.evalf(subs={x:2, y:4})
```

evaluate a function $f(x) = x^2$

### Limitations of Symbolic Differentiation

Symbolic Differentiation seems to be a great tool. But it also has some limitations. Sometimes the output expressions are too complicated and even not possible to evaluate. For example, find the derivative of the function $$\left|x\right| = \begin{cases} x, \ \text{if}\ x > 0\\  -x, \ \text{if}\ x < 0 \\ 0, \ \text{if}\ x = 0\end{cases}$$ Analytically, its derivative is:
$$\frac{d}{dx}\left(\left|x\right|\right) = \begin{cases} 1, \ \text{if}\ x > 0\\  -1, \ \text{if}\ x < 0\\\ \text{does not exist}, \ \text{if}\ x = 0\end{cases}$$

Have a look the output from the symbolic differentiation:

```python
dfdx_abs = diff(abs(x), x)
dfdx_abs
```

```python
dfdx_abs.evalf(subs={x:-2})
```

```python
dfdx_abs_numpy = lambdify(x, dfdx_abs, 'numpy')

try:
    dfdx_abs_numpy(np.array([1, -2, 0]))
except NameError as err:
    print(err)
```

You can see in this example, that you can get a very complicated function as an output of symbolic computation. This is called **expression swell**, which results in unefficiently slow computations.

## Numerical Differentiation

This method does not take into account the function expression. The only important thing is that the function can be evaluated in the nearby points $x$ and $x+\Delta x$, where $\Delta x$ is sufficiently small. Then $\frac{df}{dx}\approx\frac{f\left(x + \Delta x\right) - f\left(x\right)}{\Delta x}$, which can be called a **numerical approximation** of the derivative. 

Based on that idea there are different approaches for the numerical approximations, which somehow vary in the computation speed and accuracy. However, for all of the methods the results are not accurate - there is a round off error. At this stage there is no need to go into details of various methods, it is enough to investigate one of the numerical differentiation functions, available in `NumPy` package.

### Numerical Differentiation with `NumPy`

You can call function `np.gradient` to find the derivative of function $f\left(x\right) = x^2$ defined above. The first argument is an array of function values, the second defines the spacing $\Delta x$ for the evaluation. Here pass it as an array of $x$ values, the differences will be calculated automatically.

```python
x_array_2 = np.linspace(-5, 5, 100)
dfdx_numerical = np.gradient(f(x_array_2), x_array_2)

plot_f1_and_f2(dfdx_symb_numpy, dfdx_numerical, label1="f`(x) exact", label2="f`(x) approximate")
```

```python
# Trying out different values
x_array_b = np.linspace(-1, 5, 100)
dfdx_numerical2 = np.gradient(f(x_array_b), x_array_b)

plot_f1_and_f2(dfdx_symb_numpy, dfdx_numerical2, label1="f`(x) exact", label2="f`(x) approximate")
```

```python
# Try numerical differentiation for more complicated function:

def f_composed(x):
    return np.exp(-2*x) + 3*np.sin(3*x)

plot_f1_and_f2(lambdify(x, dfdx_composed, 'numpy'), np.gradient(f_composed(x_array_2), x_array_2),
              label1="f`(x) exact", label2="f`(x) approximate")
```

### Limitations of Numerical Differentiation
- not exact.
- its default accuracy is normally enough for ML applications.
- it is inaccurate at points where there are "jumps" of the dericative.

```python
def dfdx_abs(x):
    if x > 0:
        return 1
    else:
        if x < 0:
            return -1
        else:
            return None
        
plot_f1_and_f2(np.vectorize(dfdx_abs), np.gradient(abs(x_array_2), x_array_2))
```

- the jump .5 and -0.5 should be 1 and -1. These cause errors in computations.
- numerical differentiation is slow especially when there are hundreds of params and hundreds of derivatives being performed.

## Automatic Differentiation
- breaks down function into common functions (sin, cos, log, power functions, etc), and constructs computational graph consisting of basic functions.
- then, chain rule is used to compute the derivative at any node of the graph (very common in ML applications and neural nets).
- Disadvantage: difficulty.

Commonly used:
- `Autograd`
- `JAX`

- libraries we need: `grad`, `vmap`

### Automatic Differentiation with `JAX`

Lets try finding the derivative of the previously defined function $f(x) = x^2$ at the point $x = 3$:

```python
print("Function value at x = 3: ", f(3.0))
print("Derivative value at x = 3: ", grad(f)(3.0))
```

Remember: this can't be done with int only floats.

```python
try:
    grad(f)(3)
except TypeError as err:
    print(err)
```

Applying `grad` function to array, calculating derivative for each of its elements:

```python
try:
    grad(f)(x_array_jnp)
except TypeError as err:
    print(err)
```

This is a broadcasting issue. We can use `vmap` to solve the issue.

```python
dfdx_jax_vmap = vmap(grad(f))(x_array_jnp)
print(dfdx_jax_vmap)
```

`vmap(grad(f))` can be used to calculate the derivative of `f` for arrays of larger size. We can also plot the output:

```python
plot_f1_and_f2(f, vmap(grad(f)))
```

```python
# Lets look at all the results.

def plot_f1_and_f2(f1, f2, x_range, title):
    y_f1 = f1(x_range)
    y_f2 = f2(x_range)

    plt.figure(figsize=(12, 6))
    plt.plot(x_range, y_f1, label=f'{title} (Function)')
    plt.plot(x_range, y_f2, label=f'd/dx ({title})')
    plt.title(title)
    plt.legend()
    plt.show()

# Define the different operations in a list of functions with corresponding titles
operations = [
    (lambda x: x**3, "x**3"),
    (lambda x: 2*x**3 - 3*x**2 + 5, "2*x**3 - 3*x**2 + 5"),
    (lambda x: 1/x, "1/x"),
    (lambda x: jnp.exp(x), "exp(x)"),
    (lambda x: jnp.log(x), "log(x)"),
    (lambda x: jnp.sin(x), "sin(x)"),
    (lambda x: jnp.cos(x), "cos(x)"),
    (lambda x: jnp.abs(x), "abs(x)"),
    (lambda x: jnp.abs(x) + jnp.sin(x)*jnp.cos(x), "abs(x) + sin(x)*cos(x)")
]

# Loop through each operation, plot the function and its derivative with title
x_range = jnp.linspace(-2, 2, 100)  # Adjust range as needed for each function
for op, title in operations:
    # Here, we pass the title to the plotting function
    plot_f1_and_f2(op, vmap(grad(op)), x_range, title)
```

## Computational Efficiency of Symbolic, Numerical and Automatic Differentiation

Let's compare the speed of operations between different approaches. Let's find the derivative of a simple function $f(x) = x^2$, evaluating for an array of larger size, compare results and time used:

```python
import timeit, time
```

```python
x_array_large = np.linspace(-5, 5, 1000000)
tic_symb = time.time()
res_symb = lambdify(x, diff(f(x), x), 'numpy')(x_array_large)
toc_symb = time.time()
time_symb = 1000 * (toc_symb - tic_symb) # Time is in ms

tic_numerical = time.time()
res_numerical = np.gradient(f(x_array_large), x_array_large)
toc_numerical = time.time()
time_numerical = 1000 * (toc_numerical - tic_numerical)

tic_jax = time.time()
res_jax = vmap(grad(f))(jnp.array(x_array_large.astype('float32')))
toc_jax = time.time()
time_jax = 1000 * (toc_jax - tic_jax)

print(f"Results\nSymbolic Differentiation:\n{res_symb}\n" + 
      f"Numerical Differentiation:\n{res_numerical}\n" + 
      f"Automatic Differentiation:\n{res_jax}")

print(f"\n\nTime\nSymbolic Differentiation:\n{time_symb} ms\n" + 
      f"Numerical Differentiation:\n{time_numerical} ms\n" + 
      f"Automatic Differentiation:\n{time_jax} ms")
```

```python
def f_polynomial_simple(x):
    return 2*x**3 - 3*x**2 + 5

def f_polynomial(x):
    for i in range(3):
        x = f_polynomial_simple(x)
    return x

tic_polynomial_symb = time.time()
res_polynomial_symb = lambdify(x, diff(f_polynomial(x),x),'numpy')(x_array_large)
toc_polynomial_symb = time.time()
time_polynomial_symb = 1000 * (toc_polynomial_symb - tic_polynomial_symb)

tic_polynomial_jax = time.time()
res_polynomial_jax = vmap(grad(f_polynomial))(jnp.array(x_array_large.astype('float32')))
toc_polynomial_jax = time.time()
time_polynomial_jax = 1000 * (toc_polynomial_jax - tic_polynomial_jax)

print(f"Results\nSymbolic Differentiation:\n{res_polynomial_symb}\n" + 
      f"Automatic Differentiation:\n{res_polynomial_jax}")

print(f"\n\nTime\nSymbolic Differentiation:\n{time_polynomial_symb} ms\n" +  
      f"Automatic Differentiation:\n{time_polynomial_jax} ms")
```

- with increases in function computational graph the efficiencty of automatic differentiation raises.
- automatic differentiation uses the chain rule.
